## Análisis Avanzado con Pandas
En esta sección, cargaremos un archivo CSV, realizaremos limpieza de datos, y analizaremos estadísticas descriptivas.


```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    "Nombre": ["Alice", "Bob", "Charlie", "David"],
    "Edad": [25, 30, None, 22],
    "Salario": [50000, 60000, 70000, None],
    "Departamento": ["IT", "HR", "Finance", "IT"]
}
df = pd.DataFrame(data)

# Limpiar datos
df = df.fillna({"Edad": df["Edad"].mean(), "Salario": df["Salario"].mean()})

# Mostrar estadísticas descriptivas
print("Estadísticas descriptivas:")
print(df.describe())

# Filtrar empleados del departamento IT
it_employees = df[df["Departamento"] == "IT"]
print("\nEmpleados del departamento IT:")
print(it_employees)
```

## Visualización Avanzada
Creamos un gráfico para comparar los salarios por departamento utilizando `pandas` y `matplotlib`.


```python
import matplotlib.pyplot as plt

# Gráfico de barras: Salarios por Departamento
df.groupby("Departamento")["Salario"].mean().plot(kind="bar", color=["blue", "green", "red"])
plt.title("Salario Promedio por Departamento")
plt.xlabel("Departamento")
plt.ylabel("Salario Promedio")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.show()
```

## Decoradores en Python
Los decoradores son funciones que permiten modificar el comportamiento de otras funciones. Se utilizan comúnmente para añadir funcionalidades adicionales como:

- Medir tiempo de ejecución.
- Validar argumentos.


```python
import time

# Decorador para medir el tiempo de ejecución
def medir_tiempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fin = time.time()
        print(f"Tiempo de ejecución: {fin - inicio:.2f} segundos")
        return resultado
    return wrapper

@medir_tiempo
def calcular_suma(n):
    return sum(range(n))

# Uso del decorador
print("Suma:", calcular_suma(1000000))
```

## Trabajo con Diccionarios Anidados
Los diccionarios anidados son útiles para representar datos jerárquicos. A continuación, mostramos cómo trabajar con ellos.


```python
# Diccionario anidado
empleados = {
    "Alice": {"Edad": 25, "Salario": 50000, "Departamento": "IT"},
    "Bob": {"Edad": 30, "Salario": 60000, "Departamento": "HR"},
    "Charlie": {"Edad": 28, "Salario": 70000, "Departamento": "Finance"}
}

# Iterar sobre el diccionario
for nombre, detalles in empleados.items():
    print(f"Empleado: {nombre}")
    for clave, valor in detalles.items():
        print(f"  {clave}: {valor}")
```

## Comparación de Salarios por Edad y Departamento
Este gráfico muestra cómo varían los salarios promedio por edad en diferentes departamentos.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo
data = {
    "Edad": [22, 25, 28, 30, 35, 40],
    "IT": [40000, 45000, 50000, 55000, 60000, 65000],
    "HR": [35000, 40000, 45000, 47000, 50000, 52000],
    "Finance": [48000, 52000, 58000, 60000, 63000, 70000]
}
df = pd.DataFrame(data)

# Crear gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(df["Edad"], df["IT"], marker="o", label="IT", linestyle="--", color="blue")
plt.plot(df["Edad"], df["HR"], marker="s", label="HR", linestyle="-.", color="green")
plt.plot(df["Edad"], df["Finance"], marker="^", label="Finance", linestyle="-", color="red")

# Configurar el gráfico
plt.title("Comparación de Salarios por Edad y Departamento", fontsize=16)
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Salario Promedio", fontsize=12)
plt.legend(title="Departamento")
plt.grid(alpha=0.3)
plt.xticks(df["Edad"])
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
