```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def calcular_cuadrado(numero):
  return numero ** 2

numeros = [1, 2, 3, 4, 5]
cuadrados = [calcular_cuadrado(numero) for numero in numeros]

plt.plot(numeros, cuadrados, marker='o')
plt.xlabel('Números')
plt.ylabel('Cuadrado')
plt.title('visualizacion de Numeros al Cuadrado')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    "Nombre": ["Alice", "Bob", "Charlie", "David"],
    "Edad": [25, 30, None, 22],
    "Salario": [50000, 60000, 70000, None],
    "Departamento": ["IT", "HR", "Finance", "IT"]
}

df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
print("Estadistica del Dataframe")
print(df.describe())
```

```python
df= df.fillna({"Edad": df["Edad"].mean(), "Salario": df["Salario"].mean()})
print(df)
print("Estadistica del Dataframe")
print(df.describe())
```

```python
it_empleados= df[df["Departamento"] == "IT"]
print(it_empleados)
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras: Salarios por Departamento
df.groupby("Departamento")["Salario"].mean().plot(kind="bar", color=["blue", "green", "red"])

plt.title("Salario Promedio por Departamento")
plt.xlabel("Departamento")
plt.ylabel("Salario Promedio")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.show()
```

## Comparación de Salarios por Edad y Departamento
Este gráfico muestra cómo varían los salarios promedio por edad en diferentes departamentos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo
data = {
    "Edad": [22, 25, 28, 30, 35, 40],
    "IT": [40000, 45000, 50000, 55000, 60000, 65000],
    "HR": [35000, 40000, 45000, 47000, 50000, 52000],
    "Finance": [48000, 52000, 58000, 60000, 63000, 70000]
}
df = pd.DataFrame(data)
print(df)
try:
  data_2 = {
      "A": [1.23456, 2.34567, 3.45678],
      "B": [4.56789, 5.67891, 6.78912]
  }
  df_2 = pd.DataFrame(data_2)
  print(df_2)
  df_round = df_2.round(2)
  print( df_round)
except TypeError:
    print("Error en nombrar el tipo de dato")
except Exception as e:
    print(f"Error inesperado: {e}")
except NameError:
    print("Error de nombre")
```

```python
df.describe()
```

```python
plt.plot(df["Edad"], df["IT"], label="IT")
plt.plot(df["Edad"], df["HR"], label="HR")
plt.plot(df["Edad"], df["Finance"], label="Finance")

plt.title("Comparación de Salarios por Edad y Departamento")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.legend()
plt.grid(True)
plt.show()
```
